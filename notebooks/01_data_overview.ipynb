```python
import pandas as pd

# Load datasets from raw folder
orders      = pd.read_csv("../data/raw/olist_orders_dataset.csv")
customers   = pd.read_csv("../data/raw/olist_customers_dataset.csv")
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
payments    = pd.read_csv("../data/raw/olist_order_payments_dataset.csv")
products    = pd.read_csv("../data/raw/olist_products_dataset.csv")
sellers     = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
reviews     = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
geolocation = pd.read_csv("../data/raw/olist_geolocation_dataset.csv")
categories  = pd.read_csv("../data/raw/product_category_name_translation.csv")

# Quick check
print("Orders:", orders.shape)
print("Customers:", customers.shape)
print("Order Items:", order_items.shape)
print("Payments:", payments.shape)
```

```python
def explore_dataset(df, name):
    print(f"--- {name} ---")
    print(f"Shape: {df.shape}")
    print("\nMissing values:")
    print(df.isna().sum())
    print("\nDtypes:")
    print(df.dtypes)
    print("\nDuplicates:", df.duplicated().sum())
    print("\nPreview:")
    display(df.head())
    print("\n" + "="*80 + "\n")
```

```python
datasets = {
    "Orders": orders,
    "Customers": customers,
    "Order Items": order_items,
    "Payments": payments,
    "Products": products,
    "Sellers": sellers,
    "Reviews": reviews,
    "Geolocation": geolocation,
    "Categories": categories
}

for name, df in datasets.items():
    explore_dataset(df, name)
```

# Olist Dataset Overview

This notebook provides a high-level exploration of the raw Olist e-commerce datasets.  
We examine structure, missing values, and potential join keys.

---

## 1. Orders Dataset (`olist_orders_dataset.csv`)
- Rows: 99.441, Columns: 8
- Missing values: `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date` have nulls
- Duplicates: 0
- Key columns: `order_id`, `customer_id`
- Notes: Timestamp columns need conversion to datetime.

---

## 2. Customers Dataset (`olist_customers_dataset.csv`)
- Rows: 99,441, Columns: 5
- Missing values: None
- Duplicates: `customer_unique_id` repeats (same customer can place multiple orders)  
- Key columns: `customer_id`, `customer_unique_id`
- Notes: Each `customer_id` is unique per order

---

## 3. Order Items Dataset (`olist_order_items_dataset.csv`)
- Rows: 112,650, Columns: 7
- Missing values: None
- Duplicates: No exact duplicates, but multiple rows per `order_id` when an order has multiple products.
- Key columns: `order_id`, `product_id`, `seller_id`
- Notes: (e.g. multiple items per order)

---

## 4. Payments Dataset (`olist_order_payments_dataset.csv`)
- Rows: 103,886, Columns: 5
- Missing values: None
- Duplicates: No exact duplicates, but some orders appear multiple times (split payments).
- Key columns: `order_id`
- Notes: (e.g. multiple payments per order possible)

---

## 5. Products Dataset (`olist_products_dataset.csv`)
- Rows: 32,951, Columns: 9
- Missing values:`product_category_name` → ~6,000 nulls , `product_name_length`, `product_description_length`, `product_photos_qty` → many nulls 
- Duplicates: None
- Key columns: `product_id`
- Notes: (e.g. translation available in categories table)

---

## 6. Sellers Dataset (`olist_sellers_dataset.csv`)
- Rows: 3,100, Columns: 4
- Missing values: None
- Duplicates: None
- Key columns: `seller_id`
- Notes: (e.g. seller locations can be joined with geolocation)

---

## 7. Reviews Dataset (`olist_order_reviews_dataset.csv`)
- Rows: 100,000, Columns: 7
- Missing values:`review_comment_title` → many nulls , `review_comment_message` → many nulls  
- Duplicates: - Duplicates: Some orders may have multiple reviews. 
- Key columns: `review_id`, `order_id`
- Notes: (e.g. multiple reviews per order in rare cases)

---

## 8. Geolocation Dataset (`olist_geolocation_dataset.csv`)
- Rows: 1,000,163, Columns: 5

## 9. Categories Dataset (`product_category_name_translation.csv`)
